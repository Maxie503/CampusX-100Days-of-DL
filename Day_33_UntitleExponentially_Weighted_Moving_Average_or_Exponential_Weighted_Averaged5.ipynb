```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('/content/DailyDelhiClimateTest.csv')
```

```python
df =df.iloc[:,:2]
```

```python
plt.scatter(df['date'],df['meantemp'],c='g')
plt.plot(df['date'],df['EWMA'],label="EWMA",c='black')
plt.show()
```
