```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X_train = np.linspace(-1,1,20)
```

```python
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  ,
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475,
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test= np.linspace(-1,1,20)
```

```python
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(X_train,y_train,c='b',label='TRAIN')
plt.scatter(X_test,y_test,c='r',label='TEST')
plt.legend()
plt.show()
```

```python
model1 = Sequential()
model1.add(Dense(128,input_dim=1,activation='relu'))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='linear'))
adam = Adam(learning_rate=0.01)
model1.compile(loss='mse',metrics=['mse'],optimizer=adam)

history1 = model1.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),verbose=False)
```

```python
# evaluate the model

_, train_mse = model1.evaluate(X_train,y_train,verbose=0)
_, test_mse = model1.evaluate(X_test,y_test,verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
ypred1 = model1.predict(X_test)
```

```python
plt.scatter(X_train,y_train,c='b',label='TRAIN')
plt.scatter(X_test,y_test,c='r',label='TEST')
plt.plot(X_test,ypred1,c='g',label='PREDICTION')
plt.ylim(-1.5,1.5)
plt.legend()
plt.show()
```

```python
model2 = Sequential()
model2.add(Dense(128,input_dim=1,activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(128,activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(1,activation='linear'))
adam = Adam(learning_rate=0.01)
model2.compile(loss='mse',metrics=['mse'],optimizer=adam)

history2= model2.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),verbose=False)
```

```python
# evaluate the model

_, train_mse = model2.evaluate(X_train,y_train,verbose=0)
_, test_mse = model2.evaluate(X_test,y_test,verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
ypred2 = model2.predict(X_test)
```

```python
plt.scatter(X_train,y_train,c='b',label='TRAIN')
plt.scatter(X_test,y_test,c='r',label='TEST')
plt.plot(X_test,ypred1,c='g',label='PREDICTION')
plt.ylim(-1.5,1.5)
plt.legend()
plt.show()

plt.scatter(X_train,y_train,c='b',label='TRAIN')
plt.scatter(X_test,y_test,c='r',label='TEST')
plt.plot(X_test,ypred2,c='g',label='PREDICTION')
plt.ylim(-1.5,1.5)
plt.legend()
plt.show()
```

```python
model2 = Sequential()
model2.add(Dense(128,input_dim=1,activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(128,activation='relu'))
model2.add(Dropout(0.2))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(1,activation='linear'))
adam = Adam(learning_rate=0.01)
model2.compile(loss='mse',metrics=['mse'],optimizer=adam)

history2= model2.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test),verbose=False)

_, train_mse = model2.evaluate(X_train,y_train,verbose=0)
_, test_mse = model2.evaluate(X_test,y_test,verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))

ypred3 = model2.predict(X_test)

```
